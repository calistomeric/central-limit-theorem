### central limit theorem

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
import random
```

```python
from sklearn.datasets import load_boston
```

```python
data = load_boston()
```

```python
df_1 = pd.DataFrame(data.data, columns=data.feature_names)
```

```python
df_1
```

```python
df_target = pd.DataFrame(data.target, columns=['target'])
```

```python
df_target
```

```python
df = pd.concat([df_1, df_target], axis=1)
```

```python
df
```

```python
df_limit = df['AGE'].to_list()
```

```python
random.shuffle(df_limit)
```

```python
df_limit
```

```python
mu_pop = np.mean(df_limit) # mean of the population
```

```python
mu_pop
```

```python
std_pop = np.std(df_limit) # std of population
```

```python
std_pop
```

```python
sample_1 = df_limit[:50]
sample_2 = df_limit[60:110]
sample_3 = df_limit[120:170]
sample_4 = df_limit[180:230]
sample_5 = df_limit[240:290]
sample_6 = df_limit[300:350]
sample_7 = df_limit[360:410]
sample_8 = df_limit[420:470]
```

```python
sample_mean = [np.mean(list(sample_1)), np.mean(list(sample_2)), np.mean(list(sample_3)),\
np.mean(list(sample_4)), np.mean(list(sample_5)), np.mean(list(sample_6)), np.mean(list(sample_7))\
, np.mean(list(sample_8))]
```

```python
sample_mean
```

```python
plt.style.use('seaborn-darkgrid')
plt.figure(figsize=(10, 7))
sns.kdeplot(sample_mean)
```

```python
xbar  = np.mean(sample_mean)
xbar
```

```python
sample_list = sample_1+sample_2+sample_3+sample_4+sample_5+sample_6
```

```python
sample_std = np.std(sample_list) # std of the various population samples
```

```python
sample_std
```

for central limit theorem to be true:
1. mu_pop is approximately equal to sample mean
2. std_sample is supposed to be approximately equal to std_pop/n^1/2.... but due to the small size of the population distribution the values are well centered around the mean and thus approximately equal
